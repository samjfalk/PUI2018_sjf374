```python
histlist = []

for i in range(100):
    size = int(np.random.rand()*1990+10)
    dist = np.random.randn(size)
    
    histlist.append(dist)
    

    
```

```python
len(histlist)
```

```python
plt.figure(figsize = (10,10))

for x in histlist:
    plt.plot(len(x), x.mean(), '.')
plt.plot([0,2000], [0,0], '-')
```
