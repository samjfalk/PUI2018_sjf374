```python
samples = {}
for i in range(5000):
    samples[i] =  np.random.randn(100)
samples
```

```python
def ztest(sample, mu=0, std=1):
    xavg = sample.mean()
    n = len(sample)
    z = (mu - xavg) / (std / np.sqrt(n))
    return z
```

```python
ztest(samples[666])
```

```python
zlist = []
for k, s in samples.items():
    zlist.append(ztest(s))
```

```python
np.array(zlist).mean()
```

```python
np.array(zlist).std()
```
